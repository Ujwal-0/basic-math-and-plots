```python
x2,y2=np.meshgrid(np.linspace(-10,10,201),np.linspace(-10,10,201))
z2 = x2 + y2**2
```

```python
contourplot = plt.contour(x2,y2,z2)
plt.clabel(contourplot, inline=1, fontsize=10)
```

```python
x3, y3 = np.meshgrid(np.linspace(-10,10,201), np.linspace(-10,10,201))
z3= x3 + y3**2
plt.contourf(x3, y3, z3)
plt.colorbar()
```

```python
x4, y4 = np.meshgrid(np.linspace(-10,10,201),np.linspace(-10,10,201))
z4= np.cos(x4 + y4) + 0.05*(x4-y4)
plt.contourf(x3, y4, z4)
plt.colorbar()
```

## 3-D Plot


```python
plotproj = plt.axes(projection="3d")
#plotproj.view_init(90,90)
plotproj.contour3D(x4, y4, z4,100, cmap="binary")
```

```python
plt3d = plt.axes(projection="3d")
z = np.linspace(0,30,301)
x = np.sin(z)
y = 2*np.cos(z)
plt.xlim([-2,2])
plt.ylim([-2,2])
plt3d.plot3D(x,y,z)
```

```python
scatter3d = plt.axes(projection="3d")
z = np.linspace(0,30,301)
x = np.sin(z)
y = 2*np.cos(z)
scatter3d.scatter3D(x,y,z)
```
